```python
# 14.  How does the “in” keyword work differently for lists and dictionaries?
'''
How in Works with Lists
Checks if a value exists anywhere in the list.
Performs a linear search through all elements.
Time complexity: O(n) (slower for large lists).
How in Works with Dictionaries
Checks if a key exists in the dictionary (not values).
Uses a hash table lookup, so it's very fast.
Time complexity: O(1) average case.
'''
```

```python
# 17. Describe the time complexity of accessing elements in a dictionary?
'''
Time Complexity of Accessing Elements in a Dictionary
Accessing an element in a Python dictionary by its key typically has:
Average-case: O(1) — Constant Time
Python dictionaries use a hash table internally.
When you do value = my_dict[key], Python:
Computes the hash of the key.
Uses the hash to directly locate the position in memory.
Retrieves the value quickly without scanning the entire dictionary.
This makes lookups very fast and efficient regardless of dictionary size.
Worst-case: O(n) — Linear Time
In rare cases, hash collisions (different keys having the same hash) can occur.
Python handles collisions internally, but if many keys collide, lookup can degrade to linear search through collided entries.
However, this is uncommon due to good hash functions and resizing.
'''
```

```python
# 18.  In what situations are lists preferred over dictionaries?
'''
Lists can contain duplicates naturally. Dictionaries require unique keys.
Iterating over a list yields items in insertion order by default. This is simple and intuitive for sequences.
If you want to access elements by position (index), lists allow fast, direct indexing (my_list[2]). Dictionaries require keys.
'''
```

```python
# 19. Why are dictionaries considered unordered, and how does that affect data retrieval?
'''
Dictionaries did not preserve the insertion order of key-value pairs.
This means the order in which you added items was not guaranteed when you iterated over the dictionary or retrieved items.
Internally, dictionaries are implemented as hash tables, which focus on fast lookups rather than maintaining order.
'''
```
